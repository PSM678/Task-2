```python
def circle_area(radius):
    area = 3.14 * radius * radius
    return area

# Example
r = 5
print("Area of circle:", circle_area(r))
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Example
print("Largest:", largest(10, 25, 15))
```

```python
def string_length(s):
    count = 0
    for char in s:
        count += 1
    return count

# Example
print("Length:", string_length("Hello"))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Example
print(is_prime(7))
print(is_prime(10))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Example
print("Fibonacci:", fibonacci(6))
```

```python
def reverse_string(s):
    if s == "":
        return s
    return reverse_string(s[1:]) + s[0]

# Example
print(reverse_string("hello"))
```

```python
def sum_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Example
print("Sum:", sum_n(5))
```

```python
def count_even(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

# Example
nums = [1, 2, 3, 4, 5, 6]
print("Even count:", count_even(nums))
```

```python
for num in range(2, 101):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```
